```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.figure(figsize=(12,50))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x = '#Passengers' , y = 'Month' , data= df)
plt.title("Number of passengers throughout the months",fontsize=15, fontweight="bold")
plt.savefig('passengers.png')
```

```python
most_passengers = df[['#Passengers', 'Month']].sort_values(by="#Passengers",ascending=False)
most_passengers = most_passengers.head(10)
most_passengers
```

```python
plt.figure(figsize=(8,8))
graph=sns.barplot(y='#Passengers',x='Month', color="lightgreen",data=most_passengers)
graph.set_title('Months with the highest number of passengers')
plt.xticks(rotation=85)
plt.savefig('most_passengers.png')
plt.show()
```

```python
least_passengers = df[['#Passengers', 'Month']].sort_values(by="#Passengers",ascending=True)
least_passengers = least_passengers.head(10)
least_passengers
```

```python
plt.figure(figsize=(8,8))
graph=sns.barplot(y='#Passengers',x='Month', color="indianred",data=least_passengers)
graph.set_title('Months with the lowest number of passengers')
plt.xticks(rotation=85)
plt.savefig('least_passengers.png')
plt.show()
```

```python
first_year = df.head(12)
first_year
```

```python
plt.figure(figsize=(8,8))
graph=sns.barplot(y='#Passengers',x='Month',color='#fdae61',data=first_year)
graph.set_title('Number of passengers throughout the first year of measurement')
plt.xticks(rotation=85)
plt.savefig('first_year.png')
plt.show()
```

```python
last_year = df.tail(12)
last_year
```

```python
plt.figure(figsize=(8,8))
graph=sns.barplot(y='#Passengers',x='Month',color='aqua',data=last_year)
graph.set_title('Number of passengers throughout the last year of measurement')
plt.xticks(rotation=85)
plt.savefig('first_year.png')
plt.show()
```
